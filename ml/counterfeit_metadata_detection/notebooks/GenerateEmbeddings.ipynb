```python
import numpy as np
import pandas as pd

import pickle
```

```python
from tensorflow.keras.models import load_model

from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
df = pd.read_csv(
    "authentic_metadata_output.csv"
)

df.head()
```

```python
with open(
    "metadata_tokenizer_v1.pkl",
    "rb"
) as f:
    tokenizer=pickle.load(f)
```

```python
encoder=load_model(
    "metadata_embedding_encoder_v1.keras"
)
```

```python
metadata_texts=list(
    df["metadata"]
)
```

```python
sequences=tokenizer.texts_to_sequences(
    metadata_texts
)

max_length=24

padded_sequences=pad_sequences(
    sequences,
    maxlen=max_length,
    padding='post'
)
```

```python
metadata_embeddings=encoder.predict(
    padded_sequences
)

print(metadata_embeddings.shape)
```

```python
np.save(
    "metadata_embeddings.npy",
    metadata_embeddings
)

np.save(
    "metadata_texts.npy",
    np.array(metadata_texts)
)

np.save(
    "metadata_image_filenames.npy",
    df["image_filename"].values
)
```
